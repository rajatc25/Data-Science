```python
##########################################################################################################
##################### Dealing with numeric variables and removing outliers#################################
###########################################################################################################

################################### Check the outliers and do outlier capping##############################

plt.figure(figsize=(20, 12))
plt.subplot(3,4,1)
sns.boxplot(y = 'carlength', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,2)
sns.boxplot(y = 'carheight', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,3)
sns.boxplot(y = 'carwidth', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,4)
sns.boxplot(y = 'highwaympg', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,5)
sns.boxplot(y = 'citympg', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,6)
sns.boxplot(y = 'peakrpm', data = car_price)
# Mean and Median are close. so no outliers

plt.subplot(3,4,7)
sns.boxplot(y = 'horsepower', data = car_price)
# From boxplot we can see that there are few outliers and we need to do outliercapping
# see that 99% of data contains 207 as maximum value but if u consider 100% max value is 288. This is an outlier.
# normalise with 99%
car_price['horsepower'] = car_price['horsepower'].mask(car_price['horsepower'] >207, 207)


plt.subplot(3,4,8)
sns.boxplot(y = 'compressionratio', data = car_price)
# From boxplot we can see that there are outliers and we need to do outlier capping
# Abrupt deviation found from 91 percentile, normalise with 10.94 - 90th percentile item
car_price['compressionratio'] = car_price['compressionratio'].mask(car_price['compressionratio'] >10.94, 10.94)

plt.subplot(3,4,9)
sns.boxplot(y = 'stroke', data = car_price)
#Note that there is a jump between 0% and 1%. 
#here we will cap all values below 2.6400 to 2.6400. 
car_price['stroke']=car_price['stroke'].mask(car_price['stroke'] < 2.64, 2.64)

plt.subplot(3,4,10)
sns.boxplot(y = 'boreratio', data = car_price)
# Mean and Median are close so no outliers found.

plt.subplot(3,4,11)
sns.boxplot(y = 'curbweight', data = car_price)
# at 0% there is an outlier and above 98% there is an outlier
car_price['curbweight']=car_price['curbweight'].mask(car_price['curbweight'] > 3768.40, 3768.40)
car_price['curbweight']=car_price['curbweight'].mask(car_price['curbweight'] < 1819.72, 1819.72)


plt.subplot(3,4,12)
sns.boxplot(y = 'enginesize', data = car_price)
# There is a jump between 32% and 33%. But, we cannot have 67% outliers. 
#So, we check the next change and it is between 49% and 50% but we can not have 50% outliers
#So we check the next change and it is between 93% and 94%
#Here we will cap all values above 209.00(96% value) to 209.00
car_price['enginesize'] = car_price['enginesize'].mask(car_price['enginesize'] > 209.00, 209.00)

plt.show()
```
